```python
from os import path
import os

if os.path.exists("/content/drive/MyDrive/image_gen_mq_server/.env.txt"):
  try:
    os.rename("/content/drive/MyDrive/image_gen_mq_server/.env.txt", "/content/drive/MyDrive/image_gen_mq_server/.env")
  except FileNotFoundError:
    print("파일이 존재하지 않습니다.")
    exit(1)
  except OSError as e:
    print(f"파일 이름 변경 중 오류 발생: {e}")
    exit(1)
```
